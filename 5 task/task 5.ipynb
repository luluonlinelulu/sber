```python
import pandas as pd

class DataManager:
    def __init__(self, filepath):
        self.filepath = filepath
    
    def load_data(self):
        return pd.read_pickle(self.filepath)
    
    def save_as_csv(self, dataframe, filename):
        dataframe.to_csv(filename, index=False)
    
    def save_as_excel(self, dataframe, filename):
        dataframe.to_excel(filename, index=False)
        
if __name__ == "__main__":
    manager = DataManager('df_nashdomrf.pkl')
    df = manager.load_data()
    
    # Сохраняем данные в формате csv
    manager.save_as_csv(df, 'df_nashdomrf.csv')
    
    # Сохраняем данные в формате excel
    manager.save_as_excel(df, 'df_nashdomrf.xlsx')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

class GraphPlotter:
    def __init__(self, df):
        self.df = df

    def plot_parking_counts(self):
        grouped_data = self.df.groupby('region')
        parking_counts = grouped_data['objElemParkingCnt'].sum()

        fig, ax = plt.subplots(figsize=(17, 16))
        bars = ax.barh(parking_counts.index, parking_counts, color='#FFA500')
        ax.set_yticks(parking_counts.index)
        ax.set_yticklabels(parking_counts.index)
        ax.invert_yaxis()  # Переворачиваем ось Y
        ax.set_title('Количество парковочных мест по регионам')
        ax.set_xlabel('Количество парковочных мест')
        ax.set_ylabel('Регион')

        plt.show()

if __name__ == "__main__":
    df = pd.read_pickle('df_nashdomrf.pkl')
    graph_plotter = GraphPlotter(df)
    graph_plotter.plot_parking_counts()
```
