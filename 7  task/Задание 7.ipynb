```python
# Задание 1 ЦПТ для различных распределений


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import maxwell
import math
rv = maxwell()
sample = rv.rvs(50)
print (sample)
sample.dtype

real_mean = rv.mean() # настоящее среднее случайной величины
print ('Настоящее среднее случайной величины')
print (real_mean)


real_var = rv.var() 
print ('Настоящая дисперсия случайной величины')
print (real_var)

plt.hist(sample, density=True)
x = np.linspace(0,5,50)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='теоретическая плотность', alpha=0.5, linewidth='7')
plt.legend()

sample_mean = sample.mean() # выборочное среднее
print ('Выборочное среднее')
print(sample_mean)
sample_var = sample.var() # выборочная дисперсия
print ('Выборочная дисперсия')
print(sample_var)
Code

JupyterLab


Python 3 (ipykernel)









samplen = rv.rvs(150)

plt.hist(samplen, density=True)
x = np.linspace(0,4,150)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='теоретическая плотность', alpha=0.5, linewidth='3')
plt.legend()

samplen2 = rv.rvs(1500)

plt.hist(samplen2, density=True)
x = np.linspace(0,4,1500)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='теоретическая плотность', alpha=0.5, linewidth='1')
plt.legend()


# Для выбранного распределения апроксимация распределения выборочного среднего работает хорошо  

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts 
import statsmodels.stats.weightstats as ssw
data = pd.read_csv('C:\\Users\\Сергей\\Desktop\\telecom_churn.csv')
df = pd.DataFrame(data) 

# Данные по "неушедшим абонентам:"

mask = df['churn'].values == False
df_new = df[mask]

print("Доверительный интервал для среднего по цпт, если нам известна дисперсия: [{}, {}] ".format(
    np.mean(df_new['total day calls']) - 1.96 * np.std(df_new['total day calls'], ddof=1)/ np.sqrt(100), np.mean(df_new['total day calls']) + 1.96 * np.std(df_new['total day calls'], ddof=1)/ np.sqrt(100))) 

print("Доверительный интервал для среднего по цпт интервальная оценка: {} ".format(
sts.norm.interval(confidence=0.95, 
                  loc=np.mean(df_new['total day calls']), 
                  scale= np.std(df_new['total day calls'],                                                        
                  ddof=1)/ np.sqrt(100))))# alternative 'two-sided' only

print("Доверительный интервал для среднего по цпт методом _zconfint_generic: {} ".format(
    ssw._zconfint_generic(np.mean(df_new['total day calls']), std_mean=np.std(df_new['total day calls'], ddof=1)/np.sqrt(100), alpha=0.05, alternative='2s')
))
































```
